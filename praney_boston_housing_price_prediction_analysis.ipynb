# Exploratory Data Analysis :

# Coorelation Matrix :

```python
#creating heatmap
corr = df.corr()
plt.figure(figsize=(10,20))
sns.heatmap(corr, annot=True, cmap='Paired')
```

```python
#creaating scatterplot
plt.scatter(df['DIS'], df['NOX'])
plt.xlabel('distance from employment centre',fontsize=16)
plt.ylabel('measure of pollution',fontsize=16)
plt.title("distance v/s pollution")

```

```python
#creating histogram
plt.figure(figsize=(10,7))
sns.set(style="whitegrid")
plt.hist(df.Price,bins=70, ec="green",color="grey")
plt.xlabel("Price in thousand" , fontsize=16)
plt.ylabel("frequency")
plt.title("Price distribution")
```

```python
#creating scatterplot 
plt.scatter(df['CRIM'], df['TAX'])
plt.xlabel('per capita crime rate')
plt.ylabel('property tax tate')
plt.title('crime rate v/s tax rate')
plt.show()

```
